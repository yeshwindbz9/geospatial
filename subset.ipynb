## Reading Data Subset


```python
# Import Geopandas and Pandas
import geopandas as gpd
import pandas as pd
```

```python
# Read the Countries data (similar to a previous demo)
shp = 'data/countries.shp'
countries = gpd.read_file(shp)
countries.head()
```

```python
# Plotting
countries.plot()
```

## Reading subset of data using Geometry

```python
# Can you filter out the data from the above frame?
# For example how can we retrieve the data corresponding to only Europe?
#Let's filter using a continent

countries[countries["CONTINENT"]=="Europe"]
```

```python
# How can you plot only the filtered data?
# Plot the filtered data

countries[countries["CONTINENT"]=="Europe"].plot()
```

```python
# Can you filter the acled2019.shp data by considering only Middle East?
# Reading Acled Dataset by subseting Middle East

acled_subset = gpd.read_file("data/acled2019.shp")

acled_subset[acled_subset["region"]=="Middle East"].plot()
```

## Reading subset of data using Rows

```python
# Slice the data through inputting a range of rows
acled_subset = gpd.read_file(
    "data/acled2019.shp",
    rows=slice(100, 200)
)
acled_subset.plot();
```
